```python
#-----------------------------------Sex vs. Target (Heart Disease Outcome)--------------------------
sns.countplot(x='sex', hue='target', data=heart_df)
plt.title('Heart Disease by Sex')
plt.show()
```

```python
#-------------------# Countplot to show distribution of chest pain type by heart disease outcome---
sns.countplot(x='cp', hue='target', data=heart_df)
plt.title('Heart Disease by Chest Paint')
plt.show()
```

```python
# Histogram for Age distribution based on Heart Disease Outcome
plt.figure(figsize=(10, 6))
sns.histplot(data=heart_df, x='age', hue='target', kde=True, bins=20, palette='Set2')
plt.title('Age Distribution vs. Heart Disease Outcome')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.lineplot(x='target', y='thal', data=heart_df)
plt.title('Thal Levels vs. Heart Disease Outcome')
plt.xlabel('Heart Disease Outcome (0 = No, 1 = Yes)')
plt.ylabel('Thal Levels')
plt.show()
```

```python
sns.countplot(x='restecg', hue='target', data=heart_df)
plt.title('Resting Electrocardiographic Results vs Heart Disease Outcome')
plt.xlabel('Resting Electrocardiographic Results')
plt.ylabel('Count')
plt.show()
```
